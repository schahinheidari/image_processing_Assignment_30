```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread("input/SuperMan.jpg")
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_hsv = cv.cvtColor(img_rgb, cv.COLOR_RGB2HSV)
```

```python
H,S,V = cv.split(img_hsv)
```

```python
background = cv.imread("input/nyc.jpg")
background = cv.cvtColor(background, cv.COLOR_BGR2RGB)
plt.imshow(background)
```

```python
rows, cols = H.shape
background = cv.resize(background, (cols, rows))
```

```python
for i in range(rows):
    for j in range(cols):
        if 35 < H[i, j] < 80 and S[i, j] > 50 and V[i, j] > 60:
            img_rgb[i, j] = background[i, j]
```

```python
plt.imshow(img_rgb)
```

```python
res = cv.cvtColor(img_rgb, cv.COLOR_RGB2BGR)
cv.imwrite("output/nyc.jpg", res)
```
