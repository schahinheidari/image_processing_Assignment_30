```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread('input/potter.jpg')
img_RGB = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img_HSV = cv.cvtColor(img, cv.COLOR_BGR2HSV)
H, S, V = cv.split(img_HSV)
rows, cols, _ = img.shape
plt.imshow(img_RGB)
```

```python
blackBackground = np.zeros((rows, cols), dtype='uint8')
blackBackgroundRGB = cv.cvtColor(blackBackground, cv.COLOR_GRAY2RGB)
plt.imshow(blackBackgroundRGB)
```

```python
for i in range(rows):
    for j in range(cols):
        if 0 < H[i, j] < 15 and 40 < S[i, j] and 50 < V[i, j]:
            blackBackgroundRGB[i, j] = img_RGB[i, j]
```

```python
plt.imshow(blackBackgroundRGB)
cv.imwrite('output/potter.jpg', blackBackgroundRGB)
```
